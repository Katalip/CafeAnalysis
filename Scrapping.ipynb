```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome(executable_path='chromedriver')
driver.get('https://astana.restoran.kz/cafe?sort=shuffled')
```

```python
links = []
for i in driver.find_elements_by_class_name('place-story__title__link'):
    links.append(i.get_attribute('href'))
#cafe_data.append(driver.find_elements_by_class_name('place-story__title__link'))
```

```python
for i in range(18):
    driver.get('https://astana.restoran.kz/cafe?page=' + str(i+2) + '&sort=shuffled')
    for j in driver.find_elements_by_class_name('place-story__title__link'):
        links.append(j.get_attribute('href'))
```

```python
len(links)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns = ['Link','Name','Country','City','Street','Opening_Hours','Price_range','Cuisines','Payments','Num_of_seats','Total_n_of_views','Rating(last_year)','Num_of_reviews','Best_rating','Worst_rating','Features','Parking',
                            'For_kids','Booking','Music'])
```

```python
wait = WebDriverWait(driver, 10)
i = 0
for x in links:
    driver.get(x)
    link = 'https://astana.restoran.kz/cafe/' + x.strip('https://astana.restoran.kz/cafe/')
    
    title = None
    price_range = None
    opening_hours = None
    country = None
    city = None
    street = None
    total_n_of_views = None
    
    rating = None
    rating_count = 0
    best_rating = None
    worst_rating = None
    #reviews = ''
    features = None
    n_of_seats = None
    parking = None
    for_kids = None
    booking = None
    music = None
    
    try: 
        title = driver.find_element_by_xpath("//meta[@itemprop='name']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        price_range = driver.find_element_by_xpath("//meta[@itemprop='priceRange']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        opening_hours = driver.find_element_by_xpath("//meta[@itemprop='openingHours']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        country = driver.find_element_by_xpath("//meta[@itemprop='addressCountry']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        city = driver.find_element_by_xpath("//meta[@itemprop='addressLocality']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        street = driver.find_element_by_xpath("//meta[@itemprop='streetAddress']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        total_n_of_views = driver.find_element_by_xpath("//a[@data-target='.collapse_views_statistic']").text
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
         
    try:
        rating = driver.find_element_by_xpath("//meta[@itemprop='ratingValue']").get_attribute('content')
        rating_count = driver.find_element_by_xpath("//meta[@itemprop='ratingCount']").get_attribute('content')
        best_rating = driver.find_element_by_xpath("//meta[@itemprop='bestRating']").get_attribute('content')
        worst_rating = driver.find_element_by_xpath("//meta[@itemprop='worstRating']").get_attribute('content')
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        features = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Особенности")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        n_of_seats = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Всего мест")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        parking = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Парковка")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        for_kids = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Для детей")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        booking = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Бронирование")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    
    try: 
        music = driver.find_element_by_xpath('//tr[./td/label[contains(text(),"Музыка")]]/td[2]').text 
    except NoSuchElementException:  #spelling error making this code not work as expected
        pass
    """
    try:
        #rating = wait.until(EC.presence_of_element_located(driver.find_element_by_xpath("//meta[@itemprop='ratingValue']")))  
    except NoSuchElementException:
        pass
    """
   
    #rating_count = driver.find_element_by_xpath("//meta[@itemprop='ratingCount']").get_attribute('content')

    
    cuisines = [t.get_attribute('content') for t in (driver.find_elements_by_xpath("//meta[@itemprop='servesCuisine']"))]
    cuisines_str = ', '.join(cuisines)
    
    payments = [t.get_attribute('content') for t in (driver.find_elements_by_xpath("//meta[@itemprop='paymentAccepted']"))]
    payments_str = ', '.join(payments)
    
        
    df.loc[len(df)] = [link, title, country, city, street, opening_hours, price_range, cuisines_str, payments_str, n_of_seats, total_n_of_views , rating, rating_count, 
                       best_rating, worst_rating, features, parking, for_kids, booking, music]
    
    
```
