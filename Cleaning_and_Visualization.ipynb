```python
def process_price_range():
    global data
    prices = data['Price_range'].copy()
    l = []
    for i in prices:
        if isinstance(i,str):
            i = (int(i.split('–')[0]) + int(i.split('–')[1]))/2.0
            l.append(i)
        else:
            l.append(i)
    prices = l
    data['Price_range'] = prices
    status('Price_range')
```

### Opening_Hours

```python
data['Opening_Hours'] = data['Opening_Hours'].fillna(6)
```

```python
def process_opening_hours():
    global data
    hours = data['Opening_Hours'].copy()
    l = []
    for i in hours:
        if isinstance(i, str) and len(i.split(',')) > 1 and len(i.split(',')[0].split(' ')[0].split('–')) > 1:
            a = i.split(',')[0].split(' ')[0].split('–')[0].split(':')[0] 
            b = i.split(',')[0].split(' ')[0].split('–')[1].split(':')[0]
            if a.isdigit() and b.isdigit():
                l.append(abs(int(a)-int(b)))
            else:
                l.append(i)
        elif isinstance(i, str):
            if i.find('кр'):
                l.append(24)
            else:
                l.append(i)
        else:
            l.append(i)
            
    for i in range(293):
        if isinstance(l[i],str):
            l[i] = 24
        if l[i] < 10:
            l[i] = 10
            
    #print(l)
    data['hours_1'] = l
    #status('opening_hours')
```

```python
from collections import Counter
```

```python
def processing_cuisine():
    global data
    cuisine = data['Cuisines'].copy()
    l = []
    for i in cuisine:
        if isinstance(i,str):
            i = i.replace("'",' ')
            l.append(i.replace(',',' '))
    text = ' '.join(l)

    c = Counter()
    for line in text.splitlines():
        c.update(line.split())
    print(c.keys(),
        c.values())
    return c
```

```python
df = data.copy()
f = ['Features','For_kids','Music']
df = df.dropna(subset=f)

comment_words = ''
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df[f[0]]: 
      comment_words += val + ','
for val in df[f[1]]: 
      comment_words += val + ','
for val in df[f[2]]: 
      comment_words += val + ','
  
  
wordcloud = WordCloud(width = 1600, height = 800, 
                background_color = 'white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
```
